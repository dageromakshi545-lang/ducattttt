# Q1. Write a program to calculate the electricity bill
using only if statement? (accept number of unit from user)
according to the following criteria:

Unit Price
First 100 units no charge
Next 100 units Rs 5 per unit
After 200 units Rs 10 per unit
(For example, if input unit is 350 than total bill amount is Rs2000). Lets explain...
For first 100 no charges
Now you have 250 unit remaining
100-200 you need to pay – 100*5 = 500
Now 200-100 you need to pay 150*10 = 1500
Total = 500+1500 = 2000

```python
unit = int(input("Enter number of units: "))
bill = 0

if unit <= 100:
    bill = 0

if unit > 100 and unit <= 200:
    bill = (unit - 100) * 5

if unit > 200:
    bill = (100 * 5) + (unit - 200) * 10

print("Total electricity bill = Rs", bill)
```

# Q2. Write a program to accept percentage from the user and display
the grade according to the following criteria:

Marks Grade
> 90 A
> 80 and <= 90 B
>= 60 and <= 80 C
below 60 D

```python
per = float(input("Enter your percentage: "))

if per > 90:
    print("Grade: A")

elif per > 80 and per <= 90:
    print("Grade: B")

elif per >= 60 and per <= 80:
    print("Grade: C")

else:
    print("Grade: D")
```

# Q4. A company decided to give bonus to employee
according to following criteria:

Time period of Service Bonus

More than 10 years 10%

>=6 and <=10 8%

Less than 6 years 5%

Ask user for their salary and years of service and print the
net bonus amount?

```python
salary = float(input("Enter your salary: "))
years = int(input("Enter years of service: "))

bonus = 0

if years > 10:
    bonus = salary * 0.10

elif years >= 6 and years <= 10:
    bonus = salary * 0.08

else:
    bonus = salary * 0.05

print("Bonus amount = Rs", bonus)
```

# Q5. Accept three numbers from the user anddisplay the second largest number?

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if (a > b and a < c) or (a > c and a < b):
    print("Second largest number is:", a)

elif (b > a and b < c) or (b > c and b < a):
    print("Second largest number is:", b)

else:
    print("Second largest number is:", c)
```

# Q6. Accept the marked price from the user and calculatethe Net amount as (Marked Price – Discount) to payaccording to following criteria:

Marked Price Discount
>10000 20%
>7000 and <=10000 15%
<=7000 10%

```python
price = float(input("Enter the marked price: "))

discount = 0

if price > 10000:
    discount = price * 0.20

elif price > 7000 and price <= 10000:
    discount = price * 0.15

else:
    discount = price * 0.10

net_amount = price - discount

print("Discount =", discount)
print("Net amount to pay =", net_amount)
```
